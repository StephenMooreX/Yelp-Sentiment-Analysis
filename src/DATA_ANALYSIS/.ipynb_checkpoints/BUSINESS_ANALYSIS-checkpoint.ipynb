```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import nltk
from nltk.corpus import stopwords
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator

%matplotlib inline
```

```python
business = pd.read_csv('../input files/yelp_business.csv')
```

```python
business.head()
```

```python
f,ax = plt.subplots(1,2, figsize=(24,8))
ax1,ax2, = ax.flatten()
cnt = business['state'].value_counts()[:10].to_frame()

sns.barplot(cnt['state'], cnt.index, palette = 'RdBu', ax =ax1)
ax1.set_xlabel('')
ax1.set_title('Top 10 Areas with the Most Reviews')

cnt = business['city'].value_counts()[:10].to_frame()

sns.barplot(cnt['city'], cnt.index, palette = 'RdBu', ax =ax2)
ax2.set_xlabel('')
ax2.set_title('Top 10 Cities with the Most Reviews')
plt.subplots_adjust(wspace=0.3)
```

```python
corpus = ' '.join(business['categories'])
corpus = pd.DataFrame(corpus.split(';'), columns=['categories'])
plt.figure(figsize=(20,8))
cnt = corpus['categories'].value_counts().to_frame()[:10]
sns.barplot(cnt['categories'], cnt.index, palette = 'RdBu')
plt.title('Top 10 Reviewed Categories');
```

```python
category = ' '.join(business['categories'])
category = pd.DataFrame(corpus.split(';'), columns=['categories'])

cloud = " ".join(text for text in category.categories)
```

```python
stopwords = set(STOPWORDS)
stopwords.update(["drink", "food", "place", "go", "went"])

wordcloud = WordCloud(width=1440, 
                      height=1080, 
                      max_words=150, 
                      #stopwords=stopwords,
                      background_color="black").generate(cloud)

# Display the generated image:
# the matplotlib way:
plt.figure(figsize = (18, 13))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
```
