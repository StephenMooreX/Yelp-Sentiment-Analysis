```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import nltk
from nltk.corpus import stopwords
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator

%matplotlib inline
```

```python
reviews = pd.read_csv('../input files/yelp_review.csv', nrows= 10000)
reviews.shape
```

```python
reviews.head()
```

```python
reviews.info()
```

```python
reviews.describe()
```

```python
reviews['text length'] = reviews['text'].apply(len)
reviews.head()
```

```python
g = sns.FacetGrid(data=reviews, col='stars')
g.map(plt.hist, 'text length', bins=50)
```

```python
sns.boxplot(x='stars', y='text length', data=reviews)
```

```python
stars = reviews.groupby('stars').mean()
stars.corr()
```

```python
block=sns.heatmap(data=stars.corr(), annot=True)
```

```python
reviews['text length'] = reviews['text'].apply(len)

review_class_neg = reviews[(reviews['stars'] == 1)]
review_class_neg.shape
```

```python
text_neg = " ".join(text for text in review_class_neg.text)
print ("There are {} words total in 1 star reviews.".format(len(text_neg)))
```

```python
stopwords = set(STOPWORDS)
stopwords.update(["drink", "food", "place", "go", "went"])

wordcloud = WordCloud(width=1440, 
                      height=1080, 
                      max_words=150, 
                      stopwords=stopwords,
                      background_color="black").generate(text_neg)

# Display the generated image:
# the matplotlib way:
plt.figure(figsize = (18, 13))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
```

```python
reviews['text length'] = reviews['text'].apply(len)

review_class_pos = reviews[(reviews['stars'] == 5)]
review_class_pos.shape
```

```python
text_pos = " ".join(text for text in review_class_pos.text)
print ("There are {} words total in the 5 star reviews.".format(len(text_pos)))
```

```python
stopwords = set(STOPWORDS)
stopwords.update(["drink", "food", "place", "go", "went"])

wordcloud = WordCloud(width=1440, 
                      height=1080, 
                      max_words=150, 
                      stopwords=stopwords,
                      background_color="black").generate(text_pos)

# Display the generated image:
# the matplotlib way:
plt.figure(figsize = (18, 13))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
```
