```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator

import nltk
from nltk.corpus import stopwords
```

```python
df = pd.read_csv("../input files/yelp_review.csv", nrows= 10000)
```

```python
df.head()
```

```python
df.shape
```

```python
df['text length'] = df['text'].apply(len)

review_class_neg = df[(df['stars'] == 1)]
review_class_neg.shape
```

```python
text_neg = " ".join(text for text in review_class_neg.text)
print ("There are {} words in the negative reviews.".format(len(text_neg)))
```

```python
stopwords = set(STOPWORDS)
stopwords.update(["drink", "food", "place", "go", "went"])

wordcloud = WordCloud(stopwords=stopwords, background_color="black").generate(text_neg)

# Display the generated image:
# the matplotlib way:
plt.figure(figsize = (12, 12))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
```

```python
df['text length'] = df['text'].apply(len)

review_class_pos = df[(df['stars'] == 5)]
review_class_pos.shape
```

```python
text_pos = " ".join(text for text in review_class_pos.text)
print ("There are {} words in the negative reviews.".format(len(text_pos)))
```

```python
stopwords = set(STOPWORDS)
stopwords.update(["drink", "food", "place", "go", "went",])

wordcloud = WordCloud(stopwords=stopwords, background_color="black").generate(text_pos)

# Display the generated image:
# the matplotlib way:
plt.figure(figsize = (12, 12))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
```
